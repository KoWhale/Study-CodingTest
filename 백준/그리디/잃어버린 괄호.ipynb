```python
#잃어버린 괄호
string = input()

nums = []
ops = []

while string:
    minus = string.find('-')
    plus = string.find('+')

    if minus == -1:
        remain_nums = list(map(int, string.split('+')))
        ops += ['+'] * (len(remain_nums)-1)
        nums += remain_nums
        break
    if plus == -1:
        remain_nums = list(map(int, string.split('-')))
        ops += ['-'] * (len(remain_nums)-1)
        nums += remain_nums
        break

    if minus > plus:
        nums.append(int(string[:plus]))
        ops.append('+')
        string = string[plus+1:]
    else:
        nums.append(int(string[:minus]))
        ops.append('-')
        string = string[minus+1:]

minus_infront = False
in_bracket = 0
answer = nums[0]
for i in range(len(ops)):
    if ops[i] == '-':
        if minus_infront:
            answer -= in_bracket
            in_bracket = nums[i+1]
        else:
            minus_infront = True
            in_bracket = nums[i+1]
    elif ops[i] == '+':
        if minus_infront:
            in_bracket += nums[i+1]
        else:
            answer += nums[i+1]
            minus_infront = False

answer -= in_bracket
print(answer)
```
